```python
from PIL import Image
from scipy.ndimage import filters
import numpy.fft as fft
import numpy as np
import matplotlib.pyplot as plt
import math as math
import glob
import os
import sys
from pathlib import Path
```

```python
im_pil = Image.open(Path('imgs-P3','lena255.pgm')).convert('L')
```

```python
plt.imshow(im_pil,cmap='gray')
plt.show()
```

```python
def FT(im):
    # https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.fft.fft2.html
    return fft.fftshift(fft.fft2(im))  # perform also the shift to have lower frequencies at the center


def IFT(ft):
    return fft.ifft2(fft.ifftshift(ft))  # assumes ft is shifted and therefore reverses the shift before IFT
```

```python
 magnitude = np.log(np.absolute(FT(im_pil)))
 np.max(magnitude),np.min(magnitude)
```

```python
plt.imshow(magnitude,cmap='gray')
```

```python
plt.boxplot(magnitude.flatten(),notch=False)
plt.show()
```

```python
hist = np.histogram(magnitude.flatten(),256)[0]
plt.plot(hist)
plt.show()
```
