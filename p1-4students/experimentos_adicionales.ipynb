```python
plt.imshow(img,cmap='gray')
```

```python
f,c = img.shape
ff,cc = np.ceil(img.shape[0] / 2) ,np.ceil(img.shape[1] / 2)
a = np.mgrid[:f,:c]
a[0] = ((np.floor(a[0]/ff))%2 == 0)
a[1] = ((np.floor(a[1]/cc))%2 == 0)
a = (a[0] == a[1])
r = np.where(a,(img**2.0)/255,np.sqrt(img*255.0))
plt.imshow(r , cmap='gray')
# plt.imshow(a , cmap='gray')
```
